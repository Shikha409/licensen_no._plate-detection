```python
## for folder
from ultralytics import YOLO
#from ArabicOcr import arabicocr
import numpy as np
import cv2
import os
import easyocr
import csv

# Load the YOLO model
model = YOLO("/content/drive/MyDrive/DATA SCIENTIST_ASSIGNMENT/license_plate_detector (1).pt")

# Path to the folder containing images
folder_path = "/content/drive/MyDrive/DATA SCIENTIST_ASSIGNMENT/images"

# Create a reader object for OCR
reader = easyocr.Reader(['en'])

# Path to save the CSV file
csv_file_path = "/content/drive/MyDrive/DATA SCIENTIST_ASSIGNMENT/no.plate_results.csv"

# Open the CSV file in write mode and write the header
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Image Name", "Detected Text"])

    # Iterate over all image files in the folder
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)

        # Check if the file is an image (optional, but recommended)
        if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Predict with the model
            results = model(img_path, save=True)

            # Extract bounding boxes
            result = results[0]
            bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")

            if len(bboxes) == 0:
                print(f"No license plates detected in {img_name}.")
                detected_text = ""
            else:
                # Get the first bounding box
                x, y, x2, y2 = map(int, bboxes[0])

                # Read the image using OpenCV
                img = cv2.imread(img_path)

                # Crop the image using the bounding box
                cropped_image = img[y:y2, x:x2]

                # Read text from the cropped image
                ocr_result = reader.readtext(cropped_image)

                if len(ocr_result) == 0:
                    print(f"No text detected by OCR in {img_name}.")
                    detected_text = ""
                else:
                    # Extract the detected text from the OCR result
                    detected_text = ocr_result[0][1]
                    print(f"Detected text in {img_name}: {detected_text}")

            # Write the results to the CSV file
            writer.writerow([img_name, detected_text])

print(f"Results saved to {csv_file_path}")
```
